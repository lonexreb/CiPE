```python
def translate_pregnancy(input):
    if input == "A":
        return "Adequate and well-controlled studies have failed to demonstrate a risk to the fetus in the first trimester of pregnancy (and there is no evidence of risk in later trimesters)."
    elif input == "B":
        return "Animal reproduction studies have failed to demonstrate a risk to the fetus and there are no adequate and well-controlled studies in pregnant women."
    elif input == "C":
        return "Animal reproduction studies have shown an adverse effect on the fetus and there are no adequate and well-controlled studies in humans, but potential benefits may warrant use in pregnant women despite potential risks."
    elif input == "D":
        return "There is positive evidence of human fetal risk based on adverse reaction data from investigational or marketing experience or studies in humans, but potential benefits may warrant use in pregnant women despite potential risks."
    elif input == "X":
        return "Studies in animals or humans have demonstrated fetal abnormalities and/or there is positive evidence of human fetal risk based on adverse reaction data from investigational or marketing experience, and the risks involved in use in pregnant women clearly outweigh potential benefits."
    elif input == "N":
        return "FDA has not classified the drug."
    else:
        return "No Information Available"
```

```python
def normalize_text(text):
    # Ensure the input is a string, to handle numeric-only drug names safely
    text = str(text)
    # Remove all non-alphanumeric characters (keep numbers) and convert to lowercase
    return re.sub(r'[^\w\s]', '', text).lower()
```

```python
# Apply normalization to the 'drug' column in cohort_df and create a new column for it
cohort_df['normalized_drug'] = cohort_df['drug'].apply(normalize_text)

# Apply the same normalization to the 'generic_name' column in effect_df
effect_df['normalized_generic_name'] = effect_df['generic_name'].apply(normalize_text)
```

```python
# Perform the merge on the normalized columns
merged_df = pd.merge(cohort_df, effect_df, left_on='normalized_drug', right_on='normalized_generic_name')

# Optionally, drop the temporary normalized columns if they are no longer needed
merged_df.drop(columns=['normalized_drug', 'normalized_generic_name'], inplace=True)
```

```python
merged_df['pregnancy_category'] = merged_df['pregnancy_category'].apply(translate_pregnancy)
```
